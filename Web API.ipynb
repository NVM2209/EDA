# JOINS IN PANDAS 


```python
SQL JOINS- joins can be performed with concat,merge and join operations in pandas.
```

```python
#1. Create a first dataframe


import pandas as pd

raw_data={'subject_id':['1','2','3','4','5'],'First_name':['Alex','Ancy','Anna','Amy','Alice'],'Last_name':['Anderson','Bay','Stone','Amon','paul']}

df_a=pd.DataFrame(raw_data,columns=['subject_id','First_name','Last_name'])
df_a
```

```python
#create a second dataframe

raw_data={'subject_id':['4','5','6','7','8'],'First_name':['Billy','Boney','Bran','Batisan','Bruce'],'Last_name':['Ander','Bil','Silver','Amonly','Black']}

df_b=pd.DataFrame(raw_data,columns=['subject_id','First_name','Last_name'])
df_b
```

```python
#Union all operator

#concatenation basically glues all the df together

df_new=pd.concat([df_a,df_b],axis=0,ignore_index=True)

df_new
```

```python
#create a third dataframe

raw_data={'subject_id':['1','2','3','4','5','6','7','8'],'Test_id':[21,34,56,78,90,21,45,67]}
df_n=pd.DataFrame(raw_data,columns=['subject_id','Test_id'])
df_n
```

```python
#merger operation is used if any one column is common in df and concat operation is used when all are same in columns .




pd.merge(df_new,df_n)
```

```python
#inner joins  ---> it produce only the set records that matches in both tables.(intersection data)

pd.merge(df_a,df_b,on='subject_id',how='inner')
```

```python
# outer joins---> show all data

pd.merge(df_a,df_b,on='subject_id',how='outer')
```
